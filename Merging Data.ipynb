```python
import pandas as pd
```

```python
twitter_df = pd.read_csv("stock_sentiments.csv")
stock_df = pd.read_csv("Dow_Stock_data.csv")
stock_names = pd.read_csv("StocksDow.csv")
```

```python
twitter_df.head()
```

```python
stock_df.head()
stock_df = stock_df.rename(columns=({'Trade Date':'Date'}))
```

```python
stock_names.head()
stock_names = stock_names.rename(columns=({'Symbol':'Ticker','Company':'Stock Name'}))
stock_names.head()
```

```python
stock_merge = pd.merge(stock_df,stock_names,on='Ticker',how="left")
stock_merge.head()
```

```python
full_merge = pd.merge(twitter_df,stock_merge,on=['Stock Name',"Date"], how="left")
full_merge.head()
```

```python
full_merge.to_csv("Merged_Data.csv")
```
